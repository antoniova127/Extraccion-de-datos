**FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto
filtro1=df[df['region']=='Caribbean']
filtro1
#Use la variable de region para poder seleccionar unicamente los países del caribe.
```

```python
#Filtro por filas
filtro2= df.iloc[100:111,: ]
filtro2
#Filtrar las filas para poder explorar los paises 100 al 110 de la lista, que no son visibles cuando se despliega el dataframe.
```

```python
#Filtro por columnas
#filtro3=df.iloc[ : , 4:7]
filtro3=df.iloc[ : , [1,2,3,6]]  #columnas no consecutivas
filtro3
#Filtar la variable de columnas para poder limpiar los datos y ver más rápidamente el nombre, continente, región y la población de un país, sin tener tanta información.
```

```python
df= pd.read_csv("CountryTable.csv", index_col=2)
df
#Usar el Continente como index para poder ordenar los datos de manera ordenada y encontrar más rápido el país.
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas
filtro4=df.loc[["Africa","Europe"], ["name","population"]]
filtro4
#Usar variables de continente, nombre y población para limpiar datos y solamente mostrar el país, su población y en que continente se encuentra.
```

```python
#Filtro por cabecera
filtro5=df.head(10)
filtro5
#Se usan todas las variables, pero se usa head para mostrar únicamente los primeros 10 países de la lista.
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["population"] > 25000000]
filtro6
#Usé la variable de population para poder limpiar los datos y mostrar únicamente los países con población mayhor a 25 millones de habitantes.
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["population"] > 1000000) & (df["region"] == "Southern and Central Asia")]
filtro7
#Usé las variables de population y región para poder limpiar los datos y que aparecieran únicamente los países de la regíon central y sur de Asia con población mayor a un millón.
```

```python
#Filtro o
filtro8=df[(df["population"] > 100000000) | (df["region"] == "Southern and Central Asia")]
filtro8
#Usé las variables de population y region para poder determinar los países que están en la zona sur y central de Asia o que los países con población mayor a 10 millones.
```

```python
#Filtro not
filtro9=df[~(df["population"] >= 100000) & ~(df["region"] == "Caribbean")]
filtro9
#Usé las variables de población y región para limpiar únicamente a los datos que no tienen población mayor a 100 mil personas y que no son de la región del caribe.
```
